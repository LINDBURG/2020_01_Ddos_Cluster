```python
file_count = 0
content_count = 0
content_max = 10000

def new_file():
    global file_count
    file_count = file_count + 1
    wf = open('data_logs/data_log_' + str(file_count) + '.json', 'w', encoding = "utf-8")
    return wf

def write_one(in_file, content):
    global content_count, content_max
    content_count = content_count + 1
    
    if content_count < content_max:
        in_file.write("\t{\n")
        in_file.write('\t\t"time_start": ' + '"' + content[2] + '"\n')
        in_file.write('\t\t"source_ip": ' + '"' + content[5] + '"\n')
        in_file.write('\t\t"dest_ip": ' + '"' + content[6] + '"\n')
        in_file.write('\t\t"protocol_n": ' + '"' + content[7] + '"\n')
        in_file.write('\t\t"STRTITLE": ' + '"' + content[9] + '"\n')
        in_file.write('\t\t"source_mac": ' + '"' + content[11] + '"\n')
        in_file.write('\t\t"dest_mac": ' + '"' + content[12] + '"\n')
        in_file.write('\t\t"pkt_size": ' + '"' + content[34] + '"\n')
        in_file.write("\t},\n")
        
        return in_file
        
    elif content_count == content_max:
        in_file.write("\t{\n")
        in_file.write('\t\t"time_start": ' + '"' + content[2] + '"\n')
        in_file.write('\t\t"source_ip": ' + '"' + content[5] + '"\n')
        in_file.write('\t\t"dest_ip": ' + '"' + content[6] + '"\n')
        in_file.write('\t\t"protocol_n": ' + '"' + content[7] + '"\n')
        in_file.write('\t\t"STRTITLE": ' + '"' + content[9] + '"\n')
        in_file.write('\t\t"source_mac": ' + '"' + content[11] + '"\n')
        in_file.write('\t\t"dest_mac": ' + '"' + content[12] + '"\n')
        in_file.write('\t\t"pkt_size": ' + '"' + content[34] + '"\n')
        in_file.write("\t}\n")
        in_file.write("\n]")
        
        in_file.close()
        
        content_count = 0        
        new_f = new_file()
        new_f.write("[\n")
        return new_f

def files_make():
    file1 = open('2019-11-04 data_log.csv', 'r')
    file2 = open('2019-11-04 data_log_2.csv', 'r')
    
    wf = new_file()
    wf.write("[\n")

    rdr1 = csv.reader(file1)
    rdr2 = csv.reader(file2)
    
    header1 = next(rdr1) # 헤더 부분을 생략하고 싶은 경우
    header2 = next(rdr2)

    for content1 in rdr1:
        wf = write_one(wf, content1)

    for content2 in rdr2:
        wf = write_one(wf, content2)

    if content_count != 0:
        wf.seek(0, os.SEEK_END)
        wf.seek(wf.tell() - 3, os.SEEK_SET)
        wf.truncate()
        wf.write("\n]")

        wf.close()
        file1.close()
        file2.close()
    else:
        wf.write("\n]")
        
        wf.close()
        file1.close()
        file2.close()
```

```python
def int_to_ip(int_str):
    in_int = int(int_str)
    in_ip = socket.inet_ntoa(struct.pack('=i', in_int))
    return in_ip
```

```python
def not_match_list_add(in_pro_num):
    if in_pro_num not in protocol_not_match:
        protocol_not_match.append(in_pro_num)
```

```python
def protocol_num_match (NP):
    in_pro_num = int(NP)
    pro_str = protocol_num_dic.get(in_pro_num, "None")
    if pro_str == "None":
        not_match_list_add(in_pro_num)
    return pro_str
```
